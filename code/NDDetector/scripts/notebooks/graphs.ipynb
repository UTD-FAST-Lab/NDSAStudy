```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerTuple


df = pd.read_csv("ICSE2025_AGGREGATE_DATA.csv")
df['%consistent_result'] = df['%consistent_result'] * 100
df = df.sort_values('id')
df
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
     
     
tool_set = ['amandroid', 'flowdroid', 'infer', 'soot', 'doop', 'opal', 'code2flow', 'pycg']           
tool_name_set = ['Amandroid', 'FlowDroid', 'Infer', 'Soot', 'DOOP', 'OPAL', 'Code2flow', 'PyCG'] 

fig, axes = plt.subplots(3, 3, figsize=(6.2, 5.7), constrained_layout=True)

axes = axes.flatten()

# Loop through each tool and plot
for i, tool in enumerate(tool_set):
    df_one = df[df['tool'] == tool]
    
    # Create the boxplot and stripplot
    ax = sb.boxplot(
        data=df_one,
        patch_artist=True,
        x='configuration',
        y='%consistent_result',
        hue='tool',
        dodge=False,
        palette=["#66c2a5"],
        ax=axes[i]
    )
    sb.stripplot(
        data=df_one,
        x='configuration',
        y='%consistent_result',
        hue='tool',
        palette=["#478773"],
        dodge=False,
        jitter=False,
        size=4,
        ax=axes[i]
    )
    
    y_min = int(df_one['%consistent_result'].min())
    y_max = int(df_one['%consistent_result'].max()) + 1
    ax.set_ylim(y_min, y_max)

    ax.legend().remove()
    ax.set(xticklabels=[])
    ax.set_xlabel(f'{tool_name_set[i]} Configs', fontsize=12)
    ax.set_ylabel('% Consistent Results', fontsize=12)
    # ax.grid(True, axis='y', linestyle='--', linewidth=0.5)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])
    
plt.savefig('ICSE2025_Consistency.pdf', dpi=300, format='pdf', bbox_inches='tight')
plt
```
